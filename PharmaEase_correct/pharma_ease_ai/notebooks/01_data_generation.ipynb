# 01 Synthetic Data Generation

## Objective
Generate realistic pharmacy datasets using Faker library with 1000+ rows per dataset.

## Datasets Created
- **inventory.csv**: Drug inventory levels with usage rates and expiry dates (1500 rows)
- **sales.csv**: Transaction data with customer IDs and revenue (2000 rows)
- **prescriptions.csv**: Patient prescriptions with symptoms and recommended drugs (1300 rows)
- **employees.csv**: Employee performance metrics and attendance (1200 rows)

```python
from pathlib import Path
import numpy as np
import pandas as pd
import sys
import os

# Setup project paths - navigate from notebooks to project root
notebook_dir = Path.cwd().resolve()
project_root = notebook_dir.parent  # Go up one level from notebooks to pharma_ease_ai
data_dir = project_root / 'data'

# Create data directory if it doesn't exist
data_dir.mkdir(exist_ok=True)

sys.path.insert(0, str(project_root / 'src'))

print(f"Notebook Directory: {notebook_dir}")
print(f"Project Root: {project_root}")
print(f"Data Directory: {data_dir}")
```

```python
import random
from faker import Faker

# Initialize Faker with seed for reproducibility
fake = Faker()
random.seed(42)
np.random.seed(42)

# Define drug mappings and catalog
DRUGS = [
    "Paracetamol", "Ibuprofen", "Amoxicillin", "Metformin", "Atorvastatin",
    "Omeprazole", "Cetrizine", "Azithromycin", "Aspirin", "Losartan",
]

SYMPTOM_TO_DRUG = {
    "fever headache": "Paracetamol",
    "body pain inflammation": "Ibuprofen",
    "bacterial infection sore throat": "Amoxicillin",
    "high blood sugar fatigue": "Metformin",
    "high cholesterol chest discomfort": "Atorvastatin",
    "acid reflux stomach pain": "Omeprazole",
    "allergy sneezing itchy eyes": "Cetrizine",
    "respiratory infection cough": "Azithromycin",
    "mild pain blood thinning": "Aspirin",
    "high blood pressure dizziness": "Losartan",
}

ITEM_CATALOG = [
    "Pain Relief", "Antibiotic", "Diabetes Care", "Cardiac Care",
    "Digestive Care", "Allergy Care", "Respiratory Care", "Vitamins",
]

# Helper function for random date series
def random_date_series(n, start="2023-01-01", end="2026-03-01"):
    return pd.to_datetime(np.random.randint(
        pd.Timestamp(start).value // 10**9,
        pd.Timestamp(end).value // 10**9,
        n,
    ), unit="s").normalize()

print("Data generation engine initialized")
```

```python
# Generate Inventory Dataset
dates = random_date_series(1500)
expiry_offsets = np.random.randint(90, 1095, 1500)
inventory_data = pd.DataFrame({
    "drug_name": np.random.choice(DRUGS, 1500),
    "quantity": np.random.randint(25, 450, 1500),
    "usage_rate": np.random.uniform(1.0, 30.0, 1500).round(2),
    "date": dates,
    "expiry_date": dates + pd.to_timedelta(expiry_offsets, unit="D"),
}).sort_values("date").reset_index(drop=True)

print(f"✓ Inventory generated: {inventory_data.shape}")
```

```python
# Generate Sales Dataset
dates = random_date_series(2000)
quantity = np.random.randint(1, 8, 2000)
price = np.random.uniform(4.5, 180.0, 2000).round(2)

sales_data = pd.DataFrame({
    "transaction_id": [f"TX-{100000 + i}" for i in range(2000)],
    "items": np.random.choice(ITEM_CATALOG, 2000),
    "quantity": quantity,
    "price": price,
    "customer_id": [f"CUST-{random.randint(1000, 2500)}" for _ in range(2000)],
    "date": dates,
}).sort_values("date").reset_index(drop=True)

print(f"✓ Sales generated: {sales_data.shape}")
```

```python
# Generate Prescriptions Dataset
dates = random_date_series(1300)
symptoms = np.random.choice(list(SYMPTOM_TO_DRUG.keys()), 1300)

prescriptions_data = pd.DataFrame({
    "patient_id": [f"PAT-{random.randint(10000, 99999)}" for _ in range(1300)],
    "symptoms": symptoms,
    "drug_prescribed": [SYMPTOM_TO_DRUG[s] for s in symptoms],
    "date": dates,
}).sort_values("date").reset_index(drop=True)

print(f"✓ Prescriptions generated: {prescriptions_data.shape}")
```

```python
# Generate Employees Dataset
employees_data = pd.DataFrame({
    "employee_id": [f"EMP-{1000 + i}" for i in range(1200)],
    "shifts": np.random.choice(["Morning", "Evening", "Night"], 1200),
    "performance_score": np.random.uniform(55, 99, 1200).round(2),
    "attendance": np.random.uniform(70, 100, 1200).round(2),
    "workload": np.random.uniform(25, 100, 1200).round(2),
})

print(f"✓ Employees generated: {employees_data.shape}")
```

```python
# Save all datasets to CSV files
inventory_data.to_csv(data_dir / "inventory.csv", index=False)
sales_data.to_csv(data_dir / "sales.csv", index=False)
prescriptions_data.to_csv(data_dir / "prescriptions.csv", index=False)
employees_data.to_csv(data_dir / "employees.csv", index=False)

print("\n✓ All datasets saved to data/ directory")
```

```python
import os

# Verify dataset creation
datasets = {
    'inventory.csv': data_dir / 'inventory.csv',
    'sales.csv': data_dir / 'sales.csv',
    'prescriptions.csv': data_dir / 'prescriptions.csv',
    'employees.csv': data_dir / 'employees.csv'
}

print("\nDataset Verification:")
print("=" * 70)
for name, path in datasets.items():
    if path.exists():
        df = pd.read_csv(path)
        size_mb = os.path.getsize(path) / (1024 ** 2)
        print(f"{name:20} | Rows: {len(df):5} | Cols: {len(df.columns):2} | Size: {size_mb:.3f} MB")
    else:
        print(f"{name:20} | ❌ NOT FOUND")
```
